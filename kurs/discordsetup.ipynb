# Discord-Bots mit Python - Discord Setup

Hinweis: Diese Datei hätte genauso gut eine `.md`-Datei sein können, aber als `.ipynb` formattiert JupyterHub sie automatisch.

## Discord Developer Portal

Bevor wir Code für unseren Bot schreiben können, müssen wir ihn zuerst bei Discord erstellen. Gehe dazu zu dem [Discord Developer Portal](https://discord.com/developers/applications).

Dort kannst du, sofern du angemeldet bist, oben rechts eine neue App erstellen:

!["New Application"-Knopf oben rechts markiert](../images/discordsetup/new-application.png)

**Hinweis: Bei Discord müssen wir zwischen 'App' und 'Bot' unterscheiden: Ein Bot ist ein Teil einer App, aber die anderen Funktionen der App werden wir (fast) nicht brauchen.**

Hier solltest du einen Namen eingeben, der dir klarmacht, was deine App (oder auch dein Bot) machen soll. Keine Sorge, du kannst ihn jederzeit ändern.

Das kann, muss aber nicht, auch der Discord-Username von deinem Bot werden.

![Feld zur Eingabe vom Namen der App](../images/discordsetup/enter-application-name.png)

Jetzt bist du bei der Übersicht deiner App. Hier kannst du eine Beschreibung, Tags und ein Bild **von der App, nicht dem Bot** angeben, wenn du willst.

Danach können wir links in der Leiste 'Bot' auswählen:

![App-Übersicht, 'Bot' ist in der linken Leiste hervorgehoben](../images/discordsetup/application-overview.png)

Wie du dir wahrscheinlich denken konntest, wollen wir jetzt einen Bot hinzufügen:

![Bot-Übersicht, 'Add Bot' rechts ist hervorgehoben](../images/discordsetup/add-bot-button.png)

So kommen wir zur Bot-Übersicht:

![Bot-Übersicht](../images/discordsetup/bot-overview.png)

Hier sollte man den Benutzennamen ggfs. anpassen und ein Profilbild für den Bot hochladen.

Ich empfehle es außerdem auszustellen, dass jeder deinen Bot hinzufügen kann - mindestens bis er fertig ist.

## Deinen Bot einladen

Bots kann man nicht über normale Einladungen hinzufügen. Stattdessen brauchen wir OAuth2 - dazu müssen wir wieder zum Developer Portal.

Dort müsstest du, wenn du deine app auswählst, links in der Leiste OAuth2 > URL Generator auswählen können:

![linke Leiste beim Developer Portal, URL Generator ist hervorgehoben](../images/discordsetup/oauth2-select.png)

Dort solltest du im oberen Kasten `bot` und `application.commands` auswählen. Im unteren Kasten kannst du auswählen, welche Rechte der Bot auf deinem Server haben soll. Das kannst du nachträglich ändern - am Einfachsten ist es, dem Bot erstmal Administrator zu geben. Dann hat er zwar alle Rechte, aber auf dem neu erstellten Server gibt es nichts zu verlieren und so vermeidet man auch spätere Probleme mit den Rechten.

Jetzt müsste es so aussehen:
![URL-Generator mit bot, application.commands and administrator ausgewählt, generierte URL ist hervorgehoben](../images/discordsetup/url-generator.png)

Unten kannst du die generierte URL kopieren und in einem neuen Tab öffnen. Jetzt müsstest du deinen Test-Server auswählen und deinen Bot hinzufügen können.
